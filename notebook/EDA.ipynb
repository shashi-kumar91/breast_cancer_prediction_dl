```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore') 
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer.data, columns=cancer['feature_names'])
```

```python
df['target']=cancer['target']
```

```python
sns.histplot(df['target'])
```

```python
sns.heatmap(df.corr())
```

```python
sns.histplot(y=df['mean radius'],x=df['mean perimeter'],hue=df['target'])#above 20 mean radius and 130 perimeter most probably has breast cancer
```

```python
sns.histplot(y=df['mean smoothness'],x=df['mean perimeter'],hue=df['target'])#above 20 mean radius and 130 perimeter most probably has breast cancer
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


for column in df.columns:
    # Create the histogram plot using sns.histplot
    sns.histplot(data=df, x=column, hue="target",kde=True)  
    # Optionally, add a title to the plot based on the current column name
    plt.title(f"Histogram of {column} by Target") 
    plt.show()  # Display the plot
```
